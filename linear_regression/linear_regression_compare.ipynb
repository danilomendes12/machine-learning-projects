```python
import pandas as pd

# read data from csv to dataframe
house_pricing_data_full_train_data = pd.read_csv('data/train.csv')

#print total rows
house_pricing_data_full_train_data.info()
```

```python
import matplotlib.pyplot as plt

year_built_column_train_data = house_pricing_data_full_train_data['YearBuilt']
area_column_train_data = house_pricing_data_full_train_data['LotArea']
prices_column_train_data = house_pricing_data_full_train_data['SalePrice']

#print graph where x is year built and y is price
plt.scatter(year_built_column_train_data, prices_column_train_data)
plt.show()
```

```python
from linear_regression import linear_regression_two_features, print_result

parameters = linear_regression_two_features(year_built_column_train_data, prices_column_train_data, max_iterations=10000)
print_result(year_built_column_train_data, prices_column_train_data, parameters)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

#reshape year built column to 2d array
year_built_column_train_data = np.array(year_built_column_train_data).reshape(-1, 1)

sklearn_linear_regression = LinearRegression()
#fit the model
sklearn_linear_regression.fit(year_built_column_train_data, prices_column_train_data)
```

```python
house_pricing_data_full_test_data = pd.read_csv('data/test.csv')
year_built_test_data = house_pricing_data_full_test_data['YearBuilt']

year_built_test_data = np.array(year_built_test_data).reshape(-1, 1)
prices_prediction = sklearn_linear_regression.predict(year_built_test_data)

#print the prediction curve
plt.scatter(year_built_column_train_data, prices_column_train_data)
plt.plot(year_built_test_data, prices_prediction, color='red')
plt.show()
```
